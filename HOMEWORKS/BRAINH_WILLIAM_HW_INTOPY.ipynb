Try to answer this question using the entire dataframe : Are there any extreme sepal length values?

```python
#extreme value = STD>3.9
SPEA = df['sepal length (cm)']
STD = SPEA.std()
MEAN = SPEA.mean()

print("extreme sepal length values as below:\n")
EXTR = []
for i, j in enumerate(SPEA):
  if abs((j-MEAN)/STD) > 3.9:
    EXTR.append(i)
print(EXTR)

print("\nSeems that there is no extreme values")
```

What about other features of the flowers? Try automating the previous operation by writing a function name find_extreme_values()

```python
def find_extreme_values(df):
  for COLU in df.columns:
    SPEA = df[COLU]
    STD = SPEA.std()
    MEAN = SPEA.mean()

    print("extreme sepal in %s values as below:" % COLU)
    EXTR = []
    for i, j in enumerate(SPEA):
      if abs((j-MEAN)/STD) > 3.9:
        EXTR.append(i)
    print(EXTR)
    print("\n")

find_extreme_values(df)
```

Use this function to plot the boxplot distribution for features. Try adding a title and name for axis.

```python
df.boxplot(
    patch_artist=True,
    boxprops=dict(facecolor='lightblue', color='blue', linewidth=1.5),
    whiskerprops=dict(color='blue', linewidth=1.5),
    capprops=dict(color='blue', linewidth=1.5),
    medianprops=dict(color='red', linewidth=2),
    grid=False
)
plt.title('Iris Dataset: Features', fontsize=16, fontweight='bold')
plt.ylabel('Length (cm)', fontsize=12)
plt.xlabel('Features', fontsize=12)


plt.show()
```
